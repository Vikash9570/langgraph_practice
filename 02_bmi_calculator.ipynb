```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# 1st we define state : the variables or values we needed
class BMIState(TypedDict):
    weight_kg : float
    height_m : float
    BMI      : float
```

```python
# defining node
def calculate_bmi(state :BMIState) -> BMIState:
    weight = state['weight_kg']
    height = state['height_m']

    state['BMI'] = round(weight/(height**2),2)
    return state
```

```python
# define graph
graph = StateGraph(BMIState)

# add note to our graph(node are nothing but python function)
graph.add_node('calculate_bmi', calculate_bmi)

# add edges to our graph (start --> calculate bmi --> end)
graph.add_edge(START,'calculate_bmi')
graph.add_edge('calculate_bmi',END)

# compile the graph, return graph object
workflow = graph.compile()
```

```python
# invoke the graph
workflow.invoke({'weight_kg':80,'height_m':2})
```
